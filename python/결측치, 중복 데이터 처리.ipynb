### 1-2. 결측치 탐색 
- isnull() : 데이터프레임 및 시리즈의 결측치를 탐색해 결측치에 대해 True를 반환
- notnull() : isnull()과 반대

```python
df.head().isnull()
```

```python
df.head().notnull()
```

- sum()을 데이터프레임에 적용해 컬럼별 결측치의 수를 계산

```python
df.isnull().sum()
```

### 1-3. 결측치(누락데이터) 제거
dropna() : 결측치 제거

```python
df.dropna().info()
```

- 아무 조건이 없으면 결측치가 하나라도 존재하는 row를 아예 삭제

```python
df.dropna(axis=1).info()
```

- axis=1은 열에 대해서 적용하라는 뜻

> thresh = 500 옵션
- 결측치가 500개 이상인 열을 모두 삭제해버리라는 옵션

```python
df_thresh = df.dropna(axis=1, thresh=500)
df_thresh.info()
```

> subset = ['age'] 옵션
- 데이터프레임의 age열에 결측값이 1개라도 있으면 그 행을 drop하라는 옵션
- how='any'와 axis=0은 default값

```python
df_age = df.dropna(subset = ['age'], how = 'any', axis=0)
print(len(df_age), '/n')
print(df_age['age'].isnull().sum())
```

### 1-4. 결측치(누락 데이터) 치환
- fillna() : 결측치를 대체하는 방법

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
print("**변경 전**")
print(df['age'].head(10))

mean_age = df["age"].mean()
df['age'].fillna(mean_age, inplace=True)

print('\n')
print("**변경 후**")
print(df['age'].head(10))
```
